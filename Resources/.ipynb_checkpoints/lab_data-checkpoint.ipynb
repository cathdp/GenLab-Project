## Transform

```python
#Get total of each type of test
```

```python
def applyFunc(x):
    try:
        if x =='M1':
            return 'UMH'
        if x =='R1':
            return'Coral Springs'
        if x =='S1':
            return'UMHC'
        if x =='P1':
            return'Plantation'
        if x =='D1':
            return'Deerfield'
        if x =='K1':
            return'Kendall'
        if x =='W1':
            return'Holywood'
        if x =='L1':
            return'Lennar'
        if x =='A1':
            return'Bascom Palmer'
        if x =='V1':
            return'Aventura'
        else:
            return 'Other'
    except:
        pass
    
blood_bank['location']=blood_bank['Site'].apply(applyFunc)
```

```python
blood_bank.dropna()
blood_bank = blood_bank.rename(columns={"Test ID": "test"})
blood_bank.head()
```

```python
#Export to a CSV file
blood_bank.to_csv('./Output/blood_bank.csv', index=False)
general_data.to_csv('./Output/general_lab_data.csv', index=False)
```
